```python
import sqlite3
import ta.volatility, ta.momentum, ta.trend, ta.volume
import pandas as pd
```

```python
connection = sqlite3.connect('data/data.db')
```

```python
data = pd.read_sql('SELECT * FROM PRICES', con=connection)
data
```

```python
for s in data.symbol.unique():
    df = data[data['symbol'] == s]
    bb_indicator = ta.volatility.BollingerBands(close=df['close'], window=13, window_dev=2)
    df['bb_bbh'] = bb_indicator.bollinger_hband()
    df['bb_bbl'] = bb_indicator.bollinger_lband()

    df['rsi'] = ta.momentum.rsi(close=df['close'], window=13)

    df['roc'] = ta.momentum.roc(close=df['close'], window=13)

    df['sma26'] = ta.trend.sma_indicator(window=26)
    df['sma52'] = ta.trend.sma_indicator(window=52)

    df['vwap'] = ta.volume.volume_weighted_average_price(high=df['high'], low=df['low'], close=df['close'], volume=df['volume'], window=13)

    df['obv'] = ta.volume.on_balance_volume(close=df['close'], volume=df['volume'])

    macd_indicator = ta.trend.MACD(close=df['close'], window_slow=26, window_fast=13)
    df['macd'] = macd_indicator.macd()
    df['macd_signal'] = macd_indicator.macd_signal()
    df['macd_diff'] = macd_indicator.macd_diff()

    df['adx13'] = ta.trend.adx(high=df['high'], low=df['low'], close=df['close'], window=13)
    df['adx26'] = ta.trend.adx(high=df['high'], low=df['low'], close=df['close'], window=26)
```
